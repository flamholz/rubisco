```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn

from matplotlib import pyplot as plt
```

```python
# Markers we use for plotting below.
markers = 'o><^dsphH*'

cnames = ["windows blue", "amber", "greyish brown", "dusty purple", "dusty teal", "dusty red", ]
named_colors = seaborn.xkcd_palette(cnames)
```

```python
xs = np.logspace(0, 3, 100)
exp = -1.3
pre = 91.5
ys = np.power(xs, exp) * pre
ys_err = ys - np.random.normal(scale=0.2*ys)
ys_sub = ys_err * 0.1

figure = plt.figure(figsize=(10,10))
plt.xscale('log')
plt.yscale('log')

rand_subset = np.random.choice(xs.size, 30)
plt.scatter(xs[rand_subset], ys_err[rand_subset], c='none', edgecolors=named_colors[0], s=100)

#rand_subset = np.random.choice(xs.size, 5)
#plt.scatter(xs[rand_subset], ys_sub[rand_subset], c=named_colors[2])
plt.plot(xs, 1.5*ys, ls='--', lw=3, c=named_colors[0])
plt.fill_between(xs, 0.4*ys, 1.5*ys, color=named_colors[0], alpha=0.4)

plt.yticks([], fontsize=18)
plt.xticks([], fontsize=18)


seaborn.despine()
plt.xlabel('Parameter 1 (e.g. $k_{cat,C}$)', fontsize=24)
plt.ylabel('Parameter 2 (e.g. $K_C$)', fontsize=24)
plt.tick_params(axis=u'both', which=u'both',length=0)

plt.savefig('fig1/fig1_pareto.svg')
plt.savefig('fig1/fig1_pareto.png')


plt.show()
```
