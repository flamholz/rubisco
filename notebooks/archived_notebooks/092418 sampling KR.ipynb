```python
%matplotlib inline

import numpy as np
import power_laws
import matplotlib.colors
import seaborn
import pandas as pd
import rubisco_data

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from scipy import stats
from matplotlib import pyplot as plt
```

```python
figure = plt.figure(figsize=(10,10))

KRs = np.random.normal(10000, 2000, 1000)
for ustd in [600, 200, 20, 2]:
    dynamic_range_ratio = ustd/2000
    k11 = np.random.normal(1000, ustd, 1000)
    k5 = np.random.normal(1000, ustd, 1000)
    
    konCs = KRs*k5
    konOs = KRs*k11
    label = '$\sigma_k / \sigma_{KR} = %.2g$' % dynamic_range_ratio
    plt.loglog(konCs, konOs, 'o', label=label)
    
plt.xlabel('$v_O / K_O = k_{11} K_R$', fontsize=26)
plt.ylabel('$v_C / K_C = k_{5} K_R$', fontsize=26)
plt.legend(fontsize=16)
plt.show()
```
